```python
# Task 1: Load and Explore Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the dataset with error handling
try:
    iris = load_iris(as_frame=True)
    df = iris.frame
    
    # Save to CSV 
    df.to_csv("iris.csv", index=False)
    # Reload from CSV using pandas
    df = pd.read_csv("iris.csv")
    
    print("Dataset loaded successfully from CSV.")
except Exception as e:
    print(f"Error loading dataset: {e}")

# Display first few rows
print("\nFirst 5 rows of dataset:")
print(df.head())

# Explore dataset structure
print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

# Clean dataset (drop missing values, if any)
df = df.dropna()
print("\nDataset after handling missing values:")
print(df.head())
```

```python
# Task 2: Basic Data Analysis

# Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Grouping by species (target column is last in original iris dataset)
if "target" in df.columns:
    grouped = df.groupby("target")["petal length (cm)"].mean()
    print("\nMean petal length by species (target):")
    print(grouped)

    # Add insight
    print("\nInsight: Setosa has the smallest average petal length, "
          "while Virginica has the largest. Versicolor lies in between.")
```

```python
# Task 3: Data Visualization

# Line chart - trend of petal length across samples
plt.figure(figsize=(8,5))
plt.plot(df.index, df["petal length (cm)"], label="Petal Length")
plt.title("Line Chart: Petal Length Trend")
plt.xlabel("Sample Index")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.show()

# Bar chart - average petal length per species
plt.figure(figsize=(8,5))
sns.barplot(x="target", y="petal length (cm)", data=df, ci=None)
plt.title("Bar Chart: Avg Petal Length per Species")
plt.xlabel("Species (Target)")
plt.ylabel("Average Petal Length (cm)")
plt.show()

# Histogram - distribution of sepal length
plt.figure(figsize=(8,5))
plt.hist(df["sepal length (cm)"], bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram: Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# Scatter plot - Sepal length vs Petal length
plt.figure(figsize=(8,5))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="target", data=df)
plt.title("Scatter Plot: Sepal vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()
```

## Task 1: Load and Explore Dataset
We used the **Iris dataset**, saved it to CSV, and reloaded. 
The dataset has 150 rows and 5 columns. No missing values were found.

## Task 2: Basic Data Analysis
The descriptive statistics show that:
- Sepal length ranges from ~4.3 to 7.9 cm.
- Petal length varies the most across samples (1.0 to 6.9 cm).  

Grouping by species (target):
- **Setosa** has the smallest average petal length.
- **Virginica** has the largest.
- **Versicolor** lies in between.

## Task 3: Data Visualization
- **Line Chart:** Shows the variation in petal length across samples.
- **Bar Chart:** Confirms species-level differences in petal length.
- **Histogram:** Sepal length follows a roughly normal distribution.
- **Scatter Plot:** Strong positive relationship between sepal length and petal length, with clear species separation.
